```python
from nltk.corpus import shakespeare, gutenberg
from nltk.tokenize import sent_tokenize, PunktSentenceTokenizer
```

```python
macbeth = shakespeare.raw('macbeth.xml')
```

```python
tok = sent_tokenize(macbeth)
for i in tok[:10]:
    print(i)
    print('------------------------------------')
```

```python
poem = gutenberg.raw('blake-poems.txt')
tok2 = sent_tokenize(poem)

for i in tok2[:10]:
    print(i)
    print('--------------------')
```

```python
train_text = shakespeare.raw('j_caesar.xml')

custom_tokenizer = PunktSentenceTokenizer(train_text)

custom_tok = custom_tokenizer.tokenize(macbeth)

for i in custom_tok[:10]:
    print(i)
    print('-------------------')
```
