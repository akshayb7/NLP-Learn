```python
import nltk
from nltk.corpus import movie_reviews
import random
```

```python
random.seed(42)
```

```python
documents = [(list(movie_reviews.words(fileid)),category)
           for category in movie_reviews.categories()
           for fileid in movie_reviews.fileids(category)]
```

```python
random.shuffle(documents)
```

```python
all_words = []

for w in movie_reviews.words():
    all_words.append(w.lower())
```

```python
all_words = nltk.FreqDist(all_words)
```

```python
# Let's create a set of words which are to used as features
word_features = list(all_words.keys())[:3000]
```

```python
#len(list(all_words.keys()))
```

```python
# Function to process a document to its corresponding word features
def find_features(document):
    words = set(document)
    features ={} # Dictionary to hold True/False corresponding to presence/absence of a particular word
    for w in word_features:
        features[w] = (w in words)
    return features    
```

```python
#print(find_features(documents[0][1]))
```

```python
# Create a feature set for all the documents
featureset = [(find_features(document),category) for (document, category) in documents]
```
